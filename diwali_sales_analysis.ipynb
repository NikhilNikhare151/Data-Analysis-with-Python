```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Data Cleaning

```python
df=pd.read_csv("Diwali Sales Data.csv",encoding="unicode_escape")
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
df.info()
```

```python
# to find out null values in dataframe
pd.isnull(df).sum()
```

```python
df.shape
```

```python
#  to delete NA values in dataframe 
df.dropna(inplace=True)
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
df.info()
```

```python
df["Amount"]=df["Amount"].astype(int)
```

```python
df["Amount"].dtype
```

```python
df.columns
```

```python
df.rename(columns={"Marital_Status":"Shadi"})
```

```python
np.round(df.describe())
```

# Explonatory Data Analysis

# Gender

```python
ax=sns.countplot(x="Gender",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sales_gen
```

```python
sns.barplot(x="Gender",y="Amount",data=sales_gen)
```

from above graph we can see that most of the buyers are femails and even the purchasing power of females are greater than men

# Age

```python
df.columns
```

```python
sns.countplot(df,x="Age Group")
```

```python
sns.countplot(df,x="Age Group",hue="Gender")
```

```python
ax=sns.countplot(df,x="Age Group",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#  total amount vs age group
sales_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sales_age
```

```python
sns.barplot(sales_age,x="Age Group",y="Amount")
```

from above graphs we can see most of the buyers of age group between 26-35 years femals

<!-- State -->

```python
df.columns
```

```python
# total number of order from 10 states
sales_state=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sales_state
```

```python
plt.figure(figsize=(12,4))
sns.barplot(sales_state,x="State",y="Orders")
plt.xticks(rotation=45)
plt.title("No. of Orders from Top 10 States")
plt.show()    
```

# total amount /sales from top to states

```python
sales_state=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sales_state
```

```python
plt.figure(figsize=(12,4))
sns.barplot(sales_state,x="State",y="Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
# from above graphs we can see that unexpectabelly most of the orders are from Uttar Pradesh, Maharashtra and Karnataka 
# resppectively but total sales/amount is from UP, Karnataka and then Maharashtra
```

# Maritial Status

```python
ax=sns.countplot(data=df,x="Marital_Status")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sales_state
```

```python
sns.barplot(data=sales_state,x="Marital_Status",y="Amount",hue="Gender")
plt.show()
```

from above graphs we can see that mot of the nuyers are married(women) amd they have high purchasing power

# Occupation

```python
plt.figure(figsize=(20,5))
ax=sns.countplot(data=df,x="Occupation")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sales_state
```

```python
plt.figure(figsize=(20,5))
ax=sns.barplot(sales_state,x="Occupation",y="Amount")

```

from above graph we can see that most of the buyers are working in IT,Healthcare and Aviation Sector

# Product Category

```python
plt.figure(figsize=(20,5))
ax=sns.countplot(df,x="Product_Category")
plt.xticks(rotation=90)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sales_state
```

```python
plt.figure(figsize=(20,5))
sns.barplot(sales_state,x="Product_Category",y="Amount")
plt.xticks(rotation=90)
plt.show()
```

from above graphs we can see that most of the sold products are from Food, Clothing and Electronics Category

```python
sales_state=df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sales_state
```

```python
plt.figure(figsize=(20,5))
sns.barplot(sales_state,x="Product_ID",y="Orders")
plt.xticks()
plt.show()
```
